```python
df1 = df1.apply(lambda x: x.replace({'vw':'volvo','maxda':'mazda','toyouta':'toyota','Nissan':'nissan','porcshce':'porsche','vokswagen':'volkswagen'}, regex = True))
```

```python
df1 = df1.drop(['aspiration','doornumber','drivewheel','enginelocation'], axis = 'columns')
df1.head(1)
```

```python
df1['enginetype'].value_counts()
```

```python
df1['cylindernumber'].value_counts()
```

```python
df1 = df1.drop(['wheelbase','carlength','carwidth','carheight','curbweight','enginetype','cylindernumber','fuelsystem','boreratio','stroke','compressionratio'], axis = 'columns')
df1.head()
```

```python
df1['torque'] = ((df1['horsepower']*5250)/df1['peakrpm'])
df1.head(1)
```

```python
df1['price'] = df1['price']*82.54
```

```python
df1.head()
```

```python
df1 = df1.reindex(columns=['Company','carbody','fueltype','enginesize','horsepower','torque','citympg','highwaympg','peakrpm','price'])
df1.head()
```

```python
df1['enginesize'] = df1['enginesize'].astype(int)
df1['horsepower'] = df1['horsepower'].astype(int)
df1['citympg'] = df1['citympg'].astype(int)
df1['highwaympg'] = df1['highwaympg'].astype(int)
df1['peakrpm'] = df1['peakrpm'].astype(int)
df1['torque'] = df1['torque'].astype(int)
df1['price'] = df1['price'].astype(int)
df1.head()
```

```python
X = df1.drop('price', axis = 'columns')
X.head()
```

```python
df1['peakrpm'].value_counts()
```

```python
y=df1.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
step1 = ColumnTransformer(transformers = [
    ('col_tnf', OneHotEncoder(sparse= False,drop='first'),[0,1,2])
],remainder='passthrough')

step2 =  RandomForestRegressor(n_estimators=100,
                               random_state=3,
                               max_samples=0.5,
                               max_features=0.75,
                               max_depth=15)


pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('R2_score',r2_score(y_test,y_pred))
```
